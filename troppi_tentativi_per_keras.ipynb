```python
df = pd.read_csv('train_pronto.csv')
```

```python
df.drop(['Unnamed: 0', 'weekend'], axis = 1, inplace = True)
```

```python
class_name = 'Occupancy'
attributes = [col for col in df.columns if col != class_name]
X = df[attributes].values
y = df[class_name]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify=y)
```

```python
def build_model():
    
    n_feature = X_train.shape[1]
    model = Sequential()
    model.add(Dense(128, input_dim=n_feature, activation='relu', kernel_initializer = 'uniform'))
    model.add(Dense(64, activation='sigmoid', kernel_initializer = 'uniform'))
    model.add(Dense(32, activation = 'relu', kernel_initializer = 'uniform' ))
    model.add(Dense(16, activation = 'sigmoid', kernel_initializer = 'uniform' ))
    model.add(Dense(1, activation='tanh', kernel_initializer = 'uniform'))

    model.compile(loss='binary_crossentropy', optimizer='sgd', metrics=['accuracy'])
    return model
```

```python
model1 = build_model()

history1 = model1.fit(X_train, y_train, epochs=50, batch_size=10).history
```
